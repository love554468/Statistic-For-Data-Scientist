### Random Variables: biến ngẫu nhiên

Khái niệm biến ngẫu nhiên là nền tảng cho nhiều nội dung thống kê, thật khó để định nghĩa chính xác. Biến ngẫu nhiên là cách ánh xạ kết quả từ một tiến trình ngẫu nhiên

Như là tung đồng xu, lăn con xúc sắc, hoặc random số ta thu được 1 kết quả là 1 con số ngẫu nhiên

Tung đồng xu thì biến ngẫu nhiên X của nó có thể nhận được 1 là mặt ngửa hoặc 0 là mặt sấp

Trong ví dụ này chúng ta có tiến trình ngẫu nhiên tung đồng xu trong thí nghiệm này có thể coi là có 2 kết quả có thể xảy ra - two possible outcomes: {0,1}. Tập các khả năng gọi là tập mẫu - sample space của thí nghiệm. Mỗi lần tiến trình ngẫu nhiên được lặp lại, có có thể coi là 1 sự kiện - an event. Trong ví dụ này, tung đồng xu và lấy được kết quả sấp là một kết quả hay cũng là một sự kiện. Cơ hội và khả năng của sự kiện này xảy ra với kết quả cụ thể được gọi là probability - xác suât của sự kiện. Xác suất của sự kiện là khả năng mà biến ngẫu nhiên nhận 1 giá trị chỉ định mà được mô tả bởi P(X).
